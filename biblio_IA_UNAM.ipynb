In this notebook we explore the metadata obtained using the code get_publication.py

```python
# import the json file generated by the get_publication.py code
import json
with open('bibliography.json', 'rb') as fp:
    bibliography = json.load(fp)
n_biblio =  len(bibliography)
print(n_biblio)
```

```python
from unidecode import unidecode 
from fuzzywuzzy import fuzz
unam_aff = 'instituto de astronomia, universidad nacional autonoma de mexico'
# selection = [] 
for i in range(3):#range(n_biblio):
    selection = [] 
    list = bibliography[i]['aff']
    for aff in list:
        tmp = unidecode(aff.lower())
        selection.append(tmp)

    for aff in selection:
        print(fuzz.partial_ratio(unam_aff, aff), aff)     
```
